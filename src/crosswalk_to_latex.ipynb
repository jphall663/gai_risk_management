```python
with open(r'..' + os.sep + 'data' + os.sep + 'gai_risk_to_tc_map.pkl', 'rb') as fp:
    gai_risk_to_tc_map = pickle.load(fp)
```

```python
gai_risk_to_tc_map
```

```python
gai_risk_to_tc_frame = pd.DataFrame(columns=list(gai_risk_to_tc_map.keys()), index=range(0, 7))

for j, risk in enumerate(gai_risk_to_tc_map.keys()):
    for i, tc in enumerate(list(gai_risk_to_tc_map[risk])):  
        gai_risk_to_tc_frame.iloc[i, j] = tc
    if gai_risk_to_tc_frame[risk].isnull().any().any():
            gai_risk_to_tc_frame[risk] = gai_risk_to_tc_frame[risk].replace(np.nan, '')

gai_risk_to_tc_frame
```
