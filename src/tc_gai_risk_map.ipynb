```python
import pickle
import numpy as np
import os
import pandas as pd

# show everything
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
with open(r'..' + os.sep + 'data' + os.sep + 'data.pkl', 'rb') as fp:
    data_dict = pickle.load(fp)

data_dict['tc_gai_risk_crosswalk']
```

```python
tc_to_gai_risk_map = {}

for name in data_dict['tc_gai_risk_crosswalk'].columns:
    tc_to_gai_risk_map[name] = list(data_dict['tc_gai_risk_crosswalk'][name])
    if '' in tc_to_gai_risk_map[name]:
        tc_to_gai_risk_map[name] = [i for i in tc_to_gai_risk_map[name] if i != '']

tc_to_gai_risk_map
```

```python
with open('..' + os.sep + 'data' + os.sep + 'tc_to_gai_risk_map.pkl', 'wb') as fp:
    pickle.dump(tc_to_gai_risk_map, fp, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
gai_risks = list(data_dict['gai_risk']['Risk'].values)
gai_risk_to_tc_map = {} 

for risk in gai_risks:
    gai_risk_to_tc_map[risk] = []
    for name in list(data_dict['tc_gai_risk_crosswalk'].columns):
        if risk in list(data_dict['tc_gai_risk_crosswalk'][name]):
            gai_risk_to_tc_map[risk].append(name)
    gai_risk_to_tc_map[risk] = sorted(list(set(gai_risk_to_tc_map[risk])))
     
gai_risk_to_tc_map   
```

```python
with open('..' + os.sep + 'data' + os.sep + 'gai_risk_to_tc_map.pkl', 'wb') as fp:
    pickle.dump(gai_risk_to_tc_map, fp, protocol=pickle.HIGHEST_PROTOCOL)
```
